```python
import matplotlib.pyplot as plt
#plt.style.use('classic')
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np
import math
import pandas as pd
```

```python
#Functions
def zerolistmaker(n):
    listofzeros = [0] * n
    return listofzeros
fig = plt.figure(figsize=(20,10))
```

```python
kb=1.3806*10**(-16)
q=4.80*10**(-10)
c = 2.997925*10**10
G = 6.6732*10**(-8)
h = 6.625*10**(-27)
hbar=h/2/np.pi
eV = 1.602192*10**(-12)
me=9.109*10**(-28)
mp = 1.672614*10**(-24)
sigmaT=0.665245*10**(-24) #Thompson cross -section
#aBBeV = 7.564*10**(-15) #Radiation Density Constant eV/cm^3
#aBB = 7.564*10^(-15)/eV #Radiation Density Constant erg/cm^3
yr = 3.155815*10**7
#AU = 1.495985*10**13
pc = 3.0856*10**18
#Msun = 1.989*10**33
#Rsun = 6.96*10**10
#Lsun=3.9*10**33
#Mearth= 5.977*10^27
#Rearth = 6.37103*10**8
#r0=0.5292*10^(-8) %Aktina Bohr
```

```python
%%time
F1=2.678938534707747633656*np.pi*2**(5.0/3.0)/np.sqrt(3)
F2=np.sqrt(np.pi/2.0)
a11=-0.97947838884478688 
a12=-0.83333239129525072 
a13=0.15541796026816246 
a21=-4.69247165562628882*10**(-2.0)
a22=-0.70055018056462881
a23=1.03876297841949544*10**(-2.0)
x=np.logspace(-6.0,2.15,50)
f=[]
for i in range(len(x)):
    f.append(np.exp(a11*x[i]+a12*np.sqrt(x[i])+a13*x[i]**(1.0/3.0))*F1*x[i]**(1.0/3.0)+np.exp(a21*x[i]+a22*np.sqrt(x[i])+a23*x[i]**(1.0/3.0)-x[i])*F2*x[i]**(1.0/2.0))
plt.ylabel('$F(x)$',fontsize=20)
plt.xlabel('$x$',fontsize=20)
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
plt.loglog(x,f, linestyle='-.')  
plt.ylim(10**(-5),10**1)
```

$FLARE\ 2$
$Î’=10^3G$
$R=10^{14}cm$
$\omega_p=10(d)$
$\gamma_p=10^{5.2}$
$L'_p_{pk}=8.6 10^45$

```python
%%time
#Parameters

#CRITICAL FREQUENCY
def nu_c(gamma,B):
    return 3.0/(4.0*np.pi)*q*B/(mp*c)*gamma**2.0


def Q(gamma,t,t0,kp):
    return kp*(gamma*mp*c**2.0)**(-p)*864000**2.0/(864000**2.0+4.0*(t-t0)**2.0)


#MAGNETIC FIELD
B=10**3.0

#RADIOUS
R=10**14

#TIMESCALES
tesc=R/c
tmin=0.00
tmax=300*tesc

delta=10.0
volume=4*np.pi/3.0*R**3.0

#PROTON DISTRIBUTION
gmin=0.0
gmax=5.2
g=np.logspace(gmin,gmax,30)
p=1.5
kp=(2-p)/(-10**(gmin-p+2)+10**(gmax-p+2))/(mp*c**2.0)**(-p+2)*8.6*10**45/volume




#FREQUENCY
nu=np.logspace(2.0,np.log10(10*nu_c(g[-1],B)),50)


#INITIALIZE VALUES FOR PROTONS AND PHOTONS
Np=zerolistmaker(len(g))
photons=np.zeros(len(nu))   
a=np.zeros(len(nu))   
#maxodsed=[]


const=4.0/3.0*sigmaT*B**2.0/(8.0*np.pi*mp*c)
const1=np.sqrt(3)*q**3.0*B/(mp*c**2.0)

#FIND TIME STEP
dT=[]
for j in range (0,len(g)-1):
    dT.append(((1.0/g[j]-1.0/g[j+1]))/(4.0/3.0*sigmaT*B**2.0/(8.0*np.pi*mp*c)))
t=np.linspace(tmin,tmax,int((tmax-tmin)//dT[-1]))
dt=t[1]-t[0]
print(dt,tesc/86400)
#print('v')
print(145*10**6.0*eV/h/10**gmax)
```

```python
%%time
Np=zerolistmaker(len(g))
photons=np.zeros(len(nu))   
a=np.zeros(len(nu)) 
o=max(t)
lumx=[]
timex=[]
xx1=[]
yy1=[]
xxx1=[]
yyy1=[]
Eneutrino=[]
vLneutrino=[]

for i in range(0,int(len(t))):
    
    #g=np.append([g[0]/(g[0]*dt+1)], g)
    #Np.insert(0, 0.0)

    for j in range (0,len(g)-1):
        if t[i]<t[-1]:
            dg=g[j+1]-g[j]
            if g[j]>10**gmin:
                Np[j]=Np[j]+const*(g[j+1]**2*Np[j+1]-g[j]**2*Np[j])*dt/dg+dt*Q(g[j],t[i],int(o/2),kp)-Np[j]/tesc*dt
            else:
                Np[j]=Np[j]+const*(g[j+1]**2*Np[j+1]-g[j]**2*Np[j])*dt/dg-Np[j]/tesc*dt
            #if Np[j]<0:
                #print('H')
                #Np[j]=(Np[j+1]+Np[j-1])/2.0
                #Np[j-1]=0


        
       

    M=np.multiply(g**p,Np)
    a=0
    for l in range(len(g)-1):
        j=np.multiply(f,Np[l]*const1)
        v= np.multiply(x,nu_c(g[l],B))
        a+=np.interp(nu,v,j)*(g[l+1]-g[l])
    photons+=a*dt-photons*dt/tesc   
    #M=np.multiply(g**2,Np)
    #AA=np.multiply(nu,photons)
    if i%100==0: #or t[min(range(len(t)), key=lambda ll: abs( t[ll]-int(o/2)))]==int(o/2):
        #print(t[i]/86400)
        K=0

        for xx in range(min(range(len(nu)), key=lambda ff: abs( nu[ff]-145*10**6.0*eV/h/10**gmax)),len(nu)-1):
            
            if K==0:
                K+=4*np.pi*volume*delta**4.0*np.interp(145*10**6.0*eV/h/10**gmax,nu,a)*(nu[xx+1]-145*10**6.0*eV/h/10**gmax)
            else:
                K+=4*np.pi*volume*delta**4.0*a[xx]*(nu[xx+1]-nu[xx])
        lumx.append(K) 
        timex.append(t[i]/86400/delta)
        
        
        #Neutrino spectra
        dgdt=[]
        vL=[]
        xn=[]

        for pp in g:
            ebar=pp*nu*h

            z=0
            for m in range(min(range(len(nu)), key=lambda ff: abs( nu[ff]-145*10**6.0*eV/h/pp))-1,len(nu)-1):
                for n in range(min(range(len(nu)), key=lambda ff: abs( nu[ff]/2/h/pp-nu[m]/2/pp/h))-1,len(nu)-1):
                    z+=1.5*0.2*10**(-4)*sigmaT*ebar[m]*(ebar[m+1]-ebar[m])*photons[n]/(h*nu[n])**3.0*h*(nu[n+1]-nu[n])
            if pp*h*nu[-1]>145*10**6*1.6*10**(-12):
                xn.append(0.2/4.0*pp*mp*c**2.0*6.242*10**11) #x-axis
                dgdt.append(z)                              #Energy losses dg/dt
                vL.append(4.0*np.pi*z*c**3/pp*mp*np.interp(pp,g,Np)*delta**4.0*volume*pp*mp*c**2.0/h)
        #yy1.append(AA.copy())
        Eneutrino.append(xn.copy())
        vLneutrino.append(vL.copy())
        
        
        b=4.0*np.pi*np.multiply(a,nu)*volume*delta**4.0
        yy1.append(b.copy())
        #yy1.append(AA.copy())
        xx1.append(h*nu.copy())
        xxx1.append(g.copy())
        yyy1.append(M.copy())
```

```python
plt.figure(figsize=(20,10))
plt.title('$Flare\ 2,Î’=10^3G,R=10^{14}cm,\omega_p=10(d),\gamma_p=10^{5.2}$', fontsize=12)
plt.xlim(10**(-4.0),10**(15.0))
plt.ylim(10**(39.0),10**(46.0))

for i in range(0,len(xx1)):
    plt.loglog(6.242*10**11.0*xx1[i]*delta, (yy1[i]))
    plt.loglog(np.multiply(Eneutrino[i],delta**2.0),vLneutrino[i])
plt.ylabel('$log\ Î½L_{Î½}(erg/s)$',fontsize=20)
plt.xlabel('$log\ Îµ(eV)$',fontsize=20)
plt.figure()
plt.plot(timex,lumx)
plt.xlabel('$t(oberved)$')
plt.ylabel('$L$')
```
